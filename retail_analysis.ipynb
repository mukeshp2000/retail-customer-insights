```python
import matplotlib.pyplot as plt
```

```python
city_revenuve=(
    df.groupby("city")["amount"]
    .sum()
    .sort_values(ascending=False)
)
city_revenuve.plot(kind="bar")

plt.title("Revenue by City")    
plt.xlabel("City")
plt.ylabel("Revenue")
plt.show()  
```

```python
status_counts=df["order_status"].value_counts()
status_counts.plot(kind="pie",autopct="%1.1f%%")
plt.title("Order Status Distribution")
plt.ylabel("")
plt.show()
```

```python
customer_revenue=(df.groupby("customer_name")["amount"]
                  .sum()
                  .sort_values(ascending=False)
)   
customer_revenue.plot(kind="bar")
plt.title("Revenue by Customer")
plt.xlabel("Customer")
plt.ylabel("Revenue")
plt.show()
```
